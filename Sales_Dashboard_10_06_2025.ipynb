# **Visualization**

```python
# Group and plot
sales_Data=sales_Data.groupby('ProductCategory')['TotalAmount'].sum()
ax = sales_Data.plot(kind='bar')

# Add title and labels
plt.title('Total Sales by Product Category')
plt.ylabel('Total Sales')

# Add data labels on top of each bar
for i, value in enumerate(sales_Data):
    ax.text(i, value + 100, f'{value:.2f}', ha='center', va='bottom')

plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['ProductCategory'].value_counts()
    for x_label, grp in sales_Data.groupby('Gender')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Gender')
_ = plt.ylabel('ProductCategory')
```

```python
from matplotlib import pyplot as plt
import matplotlib.patches as patches # Import the patches module

# Create the histogram and get the axes object
ax = sales_Data['Age'].plot(kind='hist', bins=20, title='Age')

# Iterate through the patches (bars) of the histogram
for patch in ax.patches:
    # Get the height (count) and the x-position of the bar
    height = patch.get_height()
    x_pos = patch.get_x() + patch.get_width() / 2

    # Add the text label above the bar
    # f'{int(height)}' formats the label as an integer
    ax.text(x_pos, height, f'{int(height)}', ha='center', va='bottom')

plt.gca().spines[['top', 'right',]].set_visible(False)

# Display the plot
plt.show()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

# Create the horizontal bar plot and get the axes object
ax = sales_Data.groupby('ProductCategory').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))

# Iterate through the patches (bars) of the bar chart
for patch in ax.patches:
    # Get the width (count) and the y-position of the bar
    width = patch.get_width()
    y_pos = patch.get_y() + patch.get_height() / 2

    # Add the text label at the end of the bar
    # f'{int(width)}' formats the label as an integer
    ax.text(width, y_pos, f'{int(width)}', ha='left', va='center')

plt.gca().spines[['top', 'right',]].set_visible(False)

# Display the plot
plt.show()
```
